```python
import gymnasium as gym
import minigrid
from stable_baselines3 import PPO
from stable_baselines3.common.env_checker import check_env
from minigrid.wrappers import ImgObsWrapper
from minigrid_cnn import MiniGridFeaturesExtractor

env = gym.make("MiniGrid-Empty-8x8-v0", render_mode=None)

env = ImgObsWrapper(env)

print("正在检查环境兼容性...")
check_env(env)
print("环境检查通过！")

policy_kwargs = dict(
    features_extractor_class=MiniGridFeaturesExtractor,
    features_extractor_kwargs=dict(features_dim=128), # 特征维度可以调整，这里设为 128
)

model = PPO(
    "CnnPolicy", 
    env, 
    verbose=1,
    policy_kwargs=policy_kwargs # 传入定制参数
)

print("PPO 模型初始化成功！依赖库安装完美。")
```
